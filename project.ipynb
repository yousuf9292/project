```
import numpy
import pandas
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```
seed = 7
numpy.random.seed(seed)
```

```
dataframe = pandas.read_csv(path1, header=None)
dataset = dataframe.values
# split into input (X) and output (Y) variables
X = dataset[:,0:60].astype(float)
Y = dataset[:,60]
```

```
label=LabelEncoder()
label.fit(Y)
```

```
Y
```

```
Y=label.transform(Y)
```

```
Y
```

```
def create_baseline():
  model=Sequential()
  model.add(Dense(60,activation='relu',input_shape=(60,)))
  model.add(Dense(1,activation='sigmoid'))
  model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
  return model
```

```
estimator = KerasClassifier(build_fn=create_baseline, epochs=100, batch_size=5, verbose=1)
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)
results = cross_val_score(estimator, X, Y, cv=kfold)
print("Results: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```

```
scaled_X=StandardScaler().fit(X)
```

```
numpy.random.seed(seed)
estimators = []
estimators.append(('standardize', scaled_X))
estimators.append(('mlp', KerasClassifier(build_fn=create_baseline, epochs=100, batch_size=5, verbose=0)))
pipeline = Pipeline(estimators)
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)
results = cross_val_score(pipeline, X, Y, cv=kfold)
print("Standardized: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```

```
def create_smaller():
  model=Sequential()
  model.add(Dense(30,activation='relu',input_shape=(60,)))
  model.add(Dense(1,activation='sigmoid'))
  model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
  return model
```

```
numpy.random.seed(seed)
estimators = []
estimators.append(('standardize', scaled_X))
estimators.append(('mlp', KerasClassifier(build_fn=create_smaller, epochs=100, batch_size=5, verbose=1)))
pipeline = Pipeline(estimators)
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)
results = cross_val_score(pipeline, X, Y, cv=kfold)
print("Standardized: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```

```
def create_larger():
  model=Sequential()
  model.add(Dense(60,activation='relu',input_shape=(60,)))
  model.add(Dense(30,activation='relu'))
  model.add(Dense(1,activation='sigmoid'))
  model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
  return model
```

```
numpy.random.seed(seed)
estimators = []
estimators.append(('standardize', scaled_X))
estimators.append(('mlp', KerasClassifier(build_fn=create_larger, epochs=100, batch_size=5, verbose=0)))
pipeline = Pipeline(estimators)
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)
results = cross_val_score(pipeline, X, Y, cv=kfold)
print("Larger: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```

```
def create_overfit():
  model=Sequential()
  model.add(Dense(512,activation='relu',input_shape=(60,)))
  model.add(Dense(512,activation='relu'))
  model.add(Dense(128,activation='relu'))
  model.add(Dense(128,activation='relu'))
  model.add(Dense(64,activation='relu'))
  model.add(Dense(32,activation='relu'))
  model.add(Dense(28,activation='relu'))
  model.add(Dense(1,activation='sigmoid'))
  model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
  return model
```

```
numpy.random.seed(seed)
estimators = []
estimators.append(('standardize', scaled_X))  
estimators.append(('mlp', KerasClassifier(build_fn=create_overfit, epochs=150,verbose=1,validation_split=0.2)))
pipeline = Pipeline(estimators)
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)
results = cross_val_score(pipeline, X, Y, cv=kfold)
print("overfit: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```

```
def create_tune():
  model=Sequential()
  model.add(Dense(60,activation='relu',input_shape=(60,)))
  model.add(Dense(30,activation='relu'))
  model.add(Dense(1,activation='sigmoid'))
  model.compile(optimizer='rmsprop',loss='binary_crossentropy',metrics=['accuracy'])
  return model
```

```
numpy.random.seed(seed)
estimators = []
estimators.append(('standardize', scaled_X))  
estimators.append(('mlp', KerasClassifier(build_fn=create_tune, epochs=100,verbose=1,validation_split=0.1)))
pipeline = Pipeline(estimators)
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)
results = cross_val_score(pipeline, X, Y, cv=kfold)
print("tune: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```

```
from keras.models import Model
from keras.layers import Input
from keras.layers import Dense
def functional():
  visible = Input(shape=(60,))
  hidden1 = Dense(60, activation='relu')(visible)
  hidden2 = Dense(30, activation='relu')(hidden1)
  hidden3 = Dense(10, activation='relu')(hidden2)
  output = Dense(1, activation='sigmoid')(hidden3)
  model = Model(inputs=visible, outputs=output)
  model.compile(optimizer='rmsprop',loss='binary_crossentropy',metrics=['accuracy'])
  return model
```

```
numpy.random.seed(seed)
estimators = []
estimators.append(('standardize', scaled_X))  
estimators.append(('mlp', KerasClassifier(build_fn=functional, epochs=100,verbose=1)))
pipeline = Pipeline(estimators)
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)
results = cross_val_score(pipeline, X, Y, cv=kfold)
print("functional: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```

```
from keras import models
from keras import layers

def build_model():
    model = models.Sequential()
    model.add(layers.Dense(128, activation='relu',
                           input_shape=(60,)))
    model.add(layers.Dense(60, activation='relu'))
    model.add(layers.Dense(10,activation='relu'))
    model.add(layers.Dense(1))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model
model=build_model()
model.summary()
```

```
import numpy as np
numpy.random.seed(seed)
k = 2
num_val_samples = len(X) // k
num_epochs = 100
all_scores = []
for i in range(k):
    print('processing fold #', i)
    val_data = X[i * num_val_samples: (i + 1) * num_val_samples]
    val_targets = Y[i * num_val_samples: (i + 1) * num_val_samples]

    partial_train_data = np.concatenate(
        [X[:i * num_val_samples],
         X[(i + 1) * num_val_samples:]],
        axis=0)
    partial_train_targets = np.concatenate(
        [Y[:i * num_val_samples],
         Y[(i + 1) * num_val_samples:]],
        axis=0)

    model = build_model()
    model.fit(partial_train_data, partial_train_targets,
              epochs=num_epochs, batch_size=5, verbose=1)
    val_loss, val_acc = model.evaluate(val_data, val_targets, verbose=0)
    all_scores.append(val_acc)
```
